```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_table('datas/dlg375qd_9.8kn002017-8-25(16-4-31).txt')
df2 = pd.read_table('datas/dlg375qd_9.8kn012017-8-25(16-18-46).txt')
df3 = pd.read_table('datas/dlg375qd_9.8kn022017-8-25(16-36-10).txt')
```

```python
cols = ['时间', '振次', '轴位移', '力', '应力', '变形', '侧位移', '围压', '孔压', '反压', '体变', '小体变']
drop_cols = ['侧位移', '体变', '小体变', '应力']
df = pd.concat([df1,df2,df3], ignore_index=True)
df.columns = cols
df.drop(drop_cols, 1, inplace=True) #remove cols that not needed
df.dropna(inplace=True)
```

```python
df['N'] = [0.02*n for n in range(df.shape[0])]
df['d1'] = df['轴位移']-df['轴位移'][0]
df
```

```python
plt.plot(df['N'],df['d1'])
plt.xlabel('N')
plt.ylabel('d1 / mm')
plt.title('g375 9.8kn')
plt.savefig('zhouweiyi@9.8kn.png')
```
