```python
cols = ['时间', '振次', '轴位移', '力', '应力', '变形', '侧位移', '围压', '孔压', '反压', '体变', '小体变']
drop_cols = ['侧位移', '体变', '小体变', '应力']
df_14kn = pd.read_table('datas/g375dqd_14kn012017-8-29(11-28-33).txt')
df_14kn.columns = cols
df_14kn.drop(drop_cols, 1, inplace=True) #remove cols that not needed
df_14kn.dropna(inplace=True)
df_14kn['N'] = [0.02*n for n in range(df_14kn.shape[0])]
df_14kn['d1'] = df_14kn['轴位移']-df_14kn['轴位移'][0]
df_14kn['pp'] = df_14kn['孔压']-df_14kn['孔压'][0]
df_14kn.to_csv('datas/14kn.csv', index=False)
```

```python
def plotd1(df, title):
    plt.plot(df['N'],df['d1'])
    plt.xlabel('N')
    plt.ylabel('d1 / mm')
    plt.title('g375 {0}kn'.format(title))
    plt.savefig('zhouweiyi@{0}kn.png'.format(title))
```
