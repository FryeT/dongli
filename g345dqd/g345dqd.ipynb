```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import glob
files = glob.glob('datas/dqd/*.csv')
```

```python
files
```

```python
df0 = pd.read_csv(files[0], encoding='gb2312')
df1 = pd.read_csv(files[1], encoding='gb2312')
df2 = pd.read_csv(files[2], encoding='gb2312')
df3 = pd.read_csv(files[3], encoding='gb2312')
```

```python
def plotd1(df, title):
    plt.plot(df['振次'],df['epsilon_d']*100)
    plt.xlabel('N')
    plt.ylabel('d1 / %')
    plt.title('g345 {0}kn'.format(title))
```

```python
plt.figure(figsize=(8,16), dpi=300)
plt.subplot(4, 1, 1)
plotd1(df2, 5)
plt.subplot(4, 1, 2)
plotd1(df3, 8)
plt.subplot(4, 1, 3)
plotd1(df0, 12)
plt.subplot(4, 1, 4)
plotd1(df1, 15)
plt.tight_layout()
plt.savefig('g345d1.svg')
```

```python
def plotu(df, title):
    plt.plot(df['振次'],df['pore_pressure'])
    plt.xlabel('N')
    plt.ylabel('u / kPa')
    plt.title('g345 {0}kn'.format(title))
```

```python
plt.figure(figsize=(8,16), dpi=300)
plt.subplot(4, 1, 1)
plotu(df2, 5)
plt.subplot(4, 1, 2)
plotu(df3, 8)
plt.subplot(4, 1, 3)
plotu(df0, 12)
plt.subplot(4, 1, 4)
plotu(df1, 15)
plt.tight_layout()
plt.savefig('g345u.svg')
```
