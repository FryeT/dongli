```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nrows = len(files)
plt.figure(figsize=(16,20), dpi=300)
for i, f in enumerate(files):
    force = force_of_file(f)
    df = pd.read_csv(f, encoding='gb2312')
    N = df['']
    plt.subplot(nrows, 3, 3*i+1)
    plt.plot(N, df['sigma_d'])
    plt.xlabel(' N')
    plt.ylabel('$\sigma_d$ / kPa')
    plt.title('{0}kN @0.1hz'.format(force))
    
    plt.subplot(nrows, 3, 3*i+2)
    plt.plot(N, df['epsilon_d']*100)
    plt.xlabel(' N')
    plt.ylabel('$\epsilon_d$ / %')
    plt.title('{0}kN @0.1hz'.format(force))
    
    plt.subplot(nrows, 3, 3*i+3)
    plt.plot(N, df['pore_pressure'])
    plt.xlabel(' N')
    plt.ylabel('u / kPa')
    plt.title('{0}kN @0.1hz'.format(force))
plt.tight_layout()
    
```
