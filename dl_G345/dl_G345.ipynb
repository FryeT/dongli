```python
plt.figure(figsize=(12,15), dpi=300)
#plt.suptitle('G345 loops', fontsize=22)
df = pd.read_csv(g345files[1], encoding='gb2312')
plt.subplot(5, 2, 1)
plt.plot(df['epsilon_d'][:301], df['sigma_d'][:301])
plt.xlabel(r'$\varepsilon_d$', fontsize=16)
plt.ylabel(r'$\sigma_d$ / kPa', fontsize=16)
for i, f in enumerate(g345files[3:]):
    df = pd.read_csv(f, encoding='gb2312')
    plt.subplot(5, 2, i+2)
    plt.plot(df['epsilon_d'][:301], df['sigma_d'][:301])
    plt.xlabel(r'$\varepsilon_d$', fontsize=16)
    plt.ylabel(r'$\sigma_d$ / kPa', fontsize=16)
plt.tight_layout()
plt.savefig('figs/g345_loops.png')
```
