```python
g375_1_files = glob.glob('datas/0.1hz/*.csv')
g375_1_files.sort(key=force_of_file)
for i , f in enumerate(g375_1_files):
    print(i, f)
```

```python
plt.figure(figsize=(12,21), dpi=300)
#plt.suptitle('G375_1 loops', fontsize=22)
for i, f in enumerate(g375_1_files):
    df = pd.read_csv(f, encoding='gb2312')
    plt.subplot(7, 2, i+1)
    plt.plot(df['epsilon_d'], df['sigma_d'])
    plt.xlabel(r'$\varepsilon_d$', fontsize=16)
    plt.ylabel(r'$\sigma_d$ / kPa', fontsize=16)
plt.tight_layout()
plt.savefig('figs/g375_1_loops.png')
    
```
