```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_7kn = pd.read_csv('../dlg375qd/datas/7kn.csv', encoding='gb2312')
df_12kn = pd.read_csv('../dlg375qd/datas/12kn.csv', encoding='gb2312')
df_9kn = pd.read_csv('../g375dqd/datas/9kn.csv', encoding='gb2312')
df_14kn = pd.read_csv('../g375dqd/datas/14kn.csv', encoding='gb2312')
```

```python
def plotd1(df, title):
    plt.plot(df['N'],df['d1']/6)
    plt.xlabel('N', fontsize=18)
    plt.ylabel(r'$\varepsilon_z$ / %', fontsize=18)
    plt.title('g375 {0}kn'.format(title))
    
def plotu(df, title):
    plt.plot(df['N'],df['pp'])
    plt.xlabel('N', fontsize=18)
    plt.ylabel('u / kPa', fontsize=18)
    plt.title('g375 {0}kn'.format(title))
```

```python
plt.figure(figsize=(5,12.5), dpi=300)
plt.subplot(4, 1, 1)
plotu(df_7kn, 7)
plt.subplot(4, 1, 2)
plotu(df_9kn, 9.8)
plt.subplot(4, 1, 3)
plotu(df_12kn, 12)
plt.subplot(4, 1, 4)
plotu(df_14kn, 14)
plt.tight_layout()
plt.savefig('g375u.png')
```

```python
plt.figure(figsize=(5,12.5), dpi=300)
plt.subplot(4, 1, 1)
plotd1(df_7kn, 7)
plt.subplot(4, 1, 2)
plotd1(df_9kn, 9.8)
plt.subplot(4, 1, 3)
plotd1(df_12kn, 12)
plt.subplot(4, 1, 4)
plotd1(df_14kn, 14)
plt.tight_layout()
plt.savefig('g375d1.png')
```
